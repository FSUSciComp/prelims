```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
balance = np.loadtxt("balance_numpy")
balance_old = np.loadtxt("balance_numpy.old")

threads = [1, 2, 4, 8, 16]
times = np.loadtxt("times")
```

```python
plt.loglog(threads, times, marker=2)
plt.suptitle("Processor Scaling")
plt.xlabel("Number of Processors")
plt.ylabel("Wall-Clock Time (seconds)")
```

```python
plt.plot(balance, label="Updated Load Balancing")
plt.plot(balance_old, label="Naive Load Balancing")
plt.legend()
plt.title("Load Balancing")
plt.xlabel("Threads")
plt.ylabel("Per-Thread Wall-Clock Time (seconds)")
```
