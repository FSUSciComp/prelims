```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8,6)
```

## Part C - Space-Time Diagram

```python
t = np.linspace(0,2)
x = [-4, -2, -1, -0.5, 0, 0.5, 1, 2, 4]

for x0 in x:
    plt.plot(x0*np.exp(t), t)
    
axes = plt.gca()
axes.set_xlim([-8,8])
```

## Part D - Solution

```python
t = np.linspace(0,2)
x = [-4, -2, -1, -0.5, 0, 0.5, 1, 2, 4]

for x0 in x:
    plt.plot(x0*np.exp(t), t, color="lightgrey")
    
# (x tilde, 0) correspon ding to (x, t)
plt.plot(-1, 0, 'o', color='darkred')
plt.annotate(r"$(\tilde{x},0)$", (-0.8, 0), color='darkred', size=18, weight='bold')
# arbitrary (x, t)
plt.plot(-np.exp(1), 1, 'o', color='darkred')
plt.annotate(r"$(x,t)$", (-np.exp(1)+.2, 1), color='darkred', size=18, weight='bold')
# dashed line
t = np.linspace(0,1)
plt.plot(-np.exp(t), t, '--', color='darkred')

axes = plt.gca()
axes.set_xlim([-8,8])
```
