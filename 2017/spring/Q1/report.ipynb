#### (a)
$$
f(\omega, \theta)=\frac{cos\omega cos\theta + sin\omega sin\theta -1}{4(cos\omega-cos\theta)^2 + (sin\omega -sin\theta)^2}\\
=-\frac{1}{2}*\frac{(cos\omega-cos\theta)^2 + (sin\omega -sin\theta)^2}{4(cos\omega-cos\theta)^2+(sin\omega -sin\theta)^2}\\
=-\frac{1}{2}*\frac{4sin(\frac{\omega+\theta}{2})^2sin(\frac{\omega-\theta}{2})^2+4sin(\frac{\omega-\theta}{2})^2cos(\frac{\omega+\theta}{2})^2}{16sin(\frac{\omega+\theta}{2})^2sin(\frac{\omega-\theta}{2})^2+4sin(\frac{\omega-\theta}{2})^2cos(\frac{\omega+\theta}{2})^2}\\
=-\frac{1}{2}*\frac{4sin(\frac{\omega-\theta}{2})^2}{16sin(\frac{\omega+\theta}{2})^2sin(\frac{\omega-\theta}{2})^2+4sin(\frac{\omega-\theta}{2})^2cos(\frac{\omega+\theta}{2})^2}
$$
Because function $f(\omega, \theta)$ is not defined for $\omega =\theta$ (which makes the denominator 0), $4sin(\frac{\omega-\theta}{2})^2\not= 0$ we have:

$$
f(\omega, \theta)=-\frac{1}{2}*\frac{1}{4sin(\frac{\omega+\theta}{2})^2+cos(\frac{\omega+\theta}{2})^2}\\
=-\frac{1}{2}*\frac{1}{4-3cos(\frac{\omega+\theta}{2})^2}\\
$$
So we have:
$$
lim_{\omega\to\theta}=-\frac{1}{2}*\frac{1}{4-3cos\theta^2}\\
=\frac{1}{2(3cos\theta^2-4)}
$$

##### 1)
They are both periodic because:

(1)For non diagonal entries:
$$
f(\omega, (i+N)2\pi/N)=f(\omega, \theta_i+2\pi)\\
=\frac{cos\omega cos(\theta_i+2\pi)+sin\omega sin(\theta_i+2\pi)-1}{4(cos\omega-cos(\theta_i+2\pi))^2+(sin\omega-sin(\theta_i+2\pi))^2}\\
=\frac{cos \omega cos \theta_i +sin \omega sin \theta_i -1}{4 (cos \omega-cos \theta_i)^2 +(sin \omega -sin \theta_i)^2}\\
=f(\omega, \theta_i)\\
=f(\omega, i*2\pi/N)
$$

(2)For diagonal entries:
$$
f(\omega, (i+N)2\pi/N)=f(\omega, \theta_i+2\pi)\\
=\frac{1}{2(3cos(\theta_i+2\pi)^2-4)}\\
=\frac{1}{2(3cos(\theta_i)^2-4)}\\
=f(\omega, i*2\pi/N)
$$
So both of the columns are periodic with the length of period to be N. We can see form Figure 2 that the left and rigtht boundaries of the columns appears to be periodic.
