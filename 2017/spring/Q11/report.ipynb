##### 1)
for any given n, k, the likelyhood $Prob('Head'|n, k, p)$ satisfy the binomial distribution:
$$
L(p)=Prob('Head'|n,k,p)=(\frac{n}{k})p^k(1-p)^{n-k}
$$
##### 2)
$$
max_{0\le p\le 1}Prob('Head'|n,k,p)=max_{0\le p\le 1}ln[(\frac{n}{k})p^k(1-p)^{n-k}]\\
=max_{0\le p\le 1}(k ln(p)+(n-k)ln(1-p))
$$
The value of p which maximize the likelyhood function can be found when:
$$
\frac{dL(p)}{dp}=0
$$
which gives:
$$
\frac{k}{p}-\frac{n-k}{1-p}=0
$$
So we have:
$$
p_{MLE}=\frac{k}{n}
$$
In the special case of the given problem, where only one experiment done and one 'head' was got, n=1, k=1.

So:
$$
p_{MLE}=\frac{1}{1}=1
$$

##### 4)
We have the null hypothesis that the coin is fair:
$$
H_0: p=0.5
$$
The corresponding alternative hypothesis is:
$$
H_a: p\ne 0.5
$$
The likelihood ratio test was performed as following:

The likelihood ratio is given by:
$$
\Lambda=max[L(H_0|Data)]/max[L(H_a|Data)]\\
=\frac{L(p=0.5)}{L(p_{MEL})}
$$
In the cses n=k=1:
$$
\Lambda=\frac{0.5}{1}=0.5
$$
Since the two hypothesis only differ by 1 free parameter, $-2 log\Lambda$ can be compared to $\chi^2$ distribution with 1 degree of freedom.

The p value smaller than 0.05 can be reached if $\chi^2$ value is larger or equal than 3.84(from online Chi-square table), in which case the $H_0$ can be safely rejected.

We have:
$$
-2 log\Lambda=-2 log(0.5)=1.386<3.84
$$
Thus we cannot reject the null hypothesis that p=0.5, so it is still possible that the coin is fair.
