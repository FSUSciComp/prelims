#### (1)
$$
p(x)=\sum_{i=0}^n \prod_{j\ne i}^n \frac{x-x_j}{x_i-x_j}f_i\\
=\sum_{i=0}^n w_i \prod_{j\ne i}^n (x-x_j)f_i\\
=\sum_{i=0}^n w_i \frac{(x-x_0)(x-x_1)...(x-x_n)}{x-x_i}f_i\\
=(x-x_0)(x-x_1)...(x-x_n)\sum_{i=0}^{n}\frac{w_i}{x-x_i}f_i\\
=L(x)\sum_{i=0}^n \frac{w_i}{x-x_i}f_i
$$


#### (2)
Evaluate each $w_i$ for i=0, 1, ...n costs:

n substractions.

n divisions.

n-1 multiplications.

Since $w_i$ is given as:

$$
w_i=\prod_{j\ne i}^n\frac{1}{x_i-x_j}
$$
for i=0, 1,...., n

There are n+1 $w_i$ needs to be assembled(from 0 to n). The asymptotic cost in total is:

$$(n+1)*(3n-1)=O(n^2)$$
So assembling all $w_i$ needs the complexity to be $O(n^2)$

### (3)
The following algorithm can be build from equation 3 :

Given $w_i$ for i=0,1,2,...n, and the corresponding ($x_i$, $f_i$), define x as the parameter, we have:

P(x)

      result=1
    
      for i from 0 to n
        
          result=result*(x-x_i)
            
      end for
            
      temp=0
        
      for i from 0 to n
        
           temp=temp+w_i/(x-x_i)*f_i
            
      end for
            
      result=result*temp
        
      return result
      

This algorithm is O(n) in complexity since there are two for loops(not nested), each has O(n) complexity.

The above algorithm has been implemented in file lagrange.m

### (4)
To incorporate a new point should require the following steps:
#### 1)
update $w_i$:
$$
w_i= w_i \frac{1}{x_i-x_{n+1}}
$$
for i=0, 1, ....n
$$
w_{n+1}=\prod_{i=0}^n\frac{1}{x_{n+1}-x_i}
$$
This step has O(n) complexity

#### 2)
update L(x):
$$
L(x)=L(x)(x-x_{n+1})
$$
This step has O(1) complexity
#### 3)
redo the second for loop in subquestion 3, which have O(n) complexity

The total asymptotic cost is O(n) as indicated above.
