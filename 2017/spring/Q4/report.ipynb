#### (a)
Equilibriums exist when $\frac{dx}{dt}=0$ and $\frac{dy}{dt}=0$, which gives:
$$
x(r-r\frac{x}{K}-ay)=0
$$
$$
y(-b+cx)=0
$$
Depends on the parameter setting of K, there will be 2 or 3 equilibriums:
##### 1)
$$x=0$$
$$y=0$$
The biological meaning is the extinction of both species.
##### 2)
$$x=K$$
$$y=0$$
Which means the predator extinct, given that the limitation of the environmental resorces(which is proportional to K) cannot support enough preyer for maintain the survival of predator. After predator extinct, the prey start to recover in number and become saturate.
#### 3) (conditional)
$$x=\frac{b}{c}$$
$$y=\frac{r}{a}-\frac{rb}{acK}$$
This situation will only be possible when $\frac{r}{a}>\frac{rb}{acK}$, which means $K>\frac{b}{c}$, because the value of y cannot be negative, since the biological meaning of y is the density of species. If the avalable environmental resources is less than a threashold($\frac{b}{c}$), the predator will have no chance to survive. If $K>\frac{b}{c}$, both of the species will survive and maintain a certain number after osillation.

#### (b)

$$
J(x,y)=
\begin{bmatrix}
    \frac{d(\frac{dx}{dt})}{dx} & \frac{d(\frac{dx}{dt})}{dy} \\
    \frac{d(\frac{dy}{dt})}{dx} & \frac{d(\frac{dy}{dt})}{dy} \\
\end{bmatrix}
=
\begin{bmatrix}
    r-2r\frac{x}{K}-ay & -ax \\
    cy & -b+cx \\
\end{bmatrix}
$$

#### (c)
Equilibrium 1):
$$
J(0,0)=
\begin{bmatrix}
    r & 0 \\
    0 & -b \\
\end{bmatrix}
$$
The eigen values are:
$$
\lambda_1=r
$$
$$
\lambda_2=-b
$$
Equilibrium 2):
$$
J(K,0)=
\begin{bmatrix}
    -r & -aK \\
    0 & -b+cK \\
\end{bmatrix}
$$
The eigen values are:
$$
\lambda_1=-r
$$
$$
\lambda_2=-b+cK
$$

Equilibrium 3), when K>b/c:
$$
J(b/c,r/a-(rb)/(acK))=
\begin{bmatrix}
    -\frac{rb}{cK} & -\frac{ab}{c} \\
    \frac{rc}{a}-\frac{rb}{aK} & 0 \\
\end{bmatrix}
$$
The eigenvalues were calculated by symbolic tool box in MATLAB:
$$
\lambda_1=-(br+\sqrt{br(-4c^2K^2+4bcK+br)})/(2cK)
$$
$$
\lambda_2==-(br-\sqrt{br(-4c^2K^2+4bcK+br)})/(2cK)
$$
