### (3)

For the centered difference method, minimum error $\epsilon_c \approx 10^{-9} $, with the optimal step size to be $10^{-5}$

For the complex step derivative, the minimum error $\epsilon_i \approx 10^{-15} $, with the optimal step size to be less than or equal to $10^{-8}$, the accuracy will not change using a step size smaller than $10^{-8}$

### (4)

The complex step derivative will not subject to subtractive cancellation, which is a main source of round off error in centered difference formular. When calculating $f_c'(x)$, we need to use f(x+h)-f(x-h), f(x+h) and f(x-h) are both very large numbers compared to the result of f(x+h)-f(x-h), because h is a small value. Thus, the significant number will loss durning this calculation, which leads to round off error.

In centered difference approach, however, we do not have this kind of subtractive cancellation.
