#### (a)
Define $\Delta t=t^{m+1}-t^{m}$, denote the solution u corresponding to $t^m$ to be $u^m$
##### 1)
For backward Euler method:

Discrete in time:
$$
\frac{u^{m+1}-u^m}{\Delta t}=u_{xx}^{m+1}
$$
So the boundary value problem needs to be solved is:

Given $u^m$

Solve: $u^{m+1}-\Delta t u_{xx}^{m+1} =u^m$

with boundary condition: $u_x(0, t^{m+1})=u_x(2\pi, t^{m+1})$, $u(0, t^{m+1})=u(2\pi, t^{m+1})$

##### 2)
For the Crank-Nicolson scheme:

Discrete in time:
$$
\frac{u^{m+1}-u^m}{\Delta t}=(u_{xx}^{m}+u_{xx}^{m+1})/2
$$
So the boundary value problem needs to be solved is:

Given $u^m$

Solve: $\frac{u^{m+1}-u^m}{\Delta t}=(u_{xx}^{m}+u_{xx}^{m+1})/2$

with boundary condition: $u_x(0, t^{m+1})=u_x(2\pi, t^{m+1})$, $u(0, t^{m+1})=u(2\pi, t^{m+1})$

#### (b)
##### 1)
Denote "^" as fourier transform. For backward Euler:
$$u^{m+1}-\Delta t u_{xx}^{m+1} =u^m$$
We have:
$$\widehat{u^{m+1}-u^m}=\Delta t \widehat{u_{xx}^{m+1}}$$
$$\widehat{u^{m+1}_k}-\widehat{u^m_k}=\Delta t*(-k^2) \widehat{u_{k}^{m+1}}$$

where $k=-\frac{N}{2}, -\frac{N}{2}+1.... \frac{N}{2}$, for k in spectral space

So we have:
$$
\widehat{u^{m+1}_k}=\frac{1}{1+\Delta t k^2}\widehat{u^m_k}
$$

###### 2)
For Crank-Nicolson method:
$$2u^{m+1}-2u^m = \Delta t (u_{xx}^{m}+u_{xx}^{m+1})$$
$$\widehat{2u^{m+1}-2u^m} = \widehat{\Delta t (u_{xx}^{m}+u_{xx}^{m+1})}$$
So we have:
$$2\widehat{u^{m+1}_k}-2\widehat{u^m_k} = \Delta t (-k^2\widehat{u_k^{m}}-k^2\widehat{u_k^{m+1}})$$
where $k=-\frac{N}{2}, -\frac{N}{2}+1.... \frac{N}{2}$, for k in spectral space

So we have:
$$
\widehat{u^{m+1}_k}=\frac{2-\Delta t k^2}{2+\Delta t k^2}\widehat{u^m_k}
$$

#### (c)
To guarantee the smoothness of the method, the function $\bigwedge$ should satisfy:

$$|\Lambda(k, \Delta t)|\le 1$$

##### 1)
For backward Euler method:
$$|\Lambda(k, \Delta t)|= | \frac{1}{1+\Delta t k^2}|$$
Which is guaranteed to be less than 1, given $\Delta t>0, k^2 \ge 0$. So the vale of $\widehat{u_k}$ will reduce as a function of time for each k in z. The solution will become smooth as a function of time.

For the high frequency components $\widehat{u_k}$ with large k, $|\Lambda(k, \Delta t)|$ will be smaller as the k becomes larger, thus the smoothing effect on the high frequency components will be even more significant, since the $\widehat{u_k}$ will reduce faser compared with lower frequency components with smaller value of k.

##### 2)
For Crank-Nicolson method:
$$|\Lambda(k, \Delta t)|= | \frac{2-\Delta t k^2}{2+\Delta t k^2}|$$, which is also guaranteed to be less than 1, given $\Delta t>0, k^2 \ge 0$. so the value of $\widehat{u_k}$ will reduce as a function of time, and the solution will become smooth correspondingly.

For the high frequency components, the sitoation is different from the backward Euler method: The  $|\Lambda(k, \Delta t)|$ will reach the minimum when $k=\sqrt{\frac{2}{\Delta t}}$. $|\Lambda|$ increases to 1 from below as |k| approximate to positive infitity. So the increasing of  smoothness of the frequency component with k greater than $\sqrt{\frac{2}{\Delta t}}$ will become less significant as k becomes larger.
