```matlab
%Prelim Question 4 Part 3
%Plot the given probability distribution
% Please input standard deviation as stan_dev
stan_dev = 1;
y0_distrib = zeros(100,1);
y1_distrib = zeros(100,1);
x = zeros(100,1);
for j = 1:100
    x(j,1) = (0.2)*(j) - (10) ;
      
end
for i = 1:100 
xx = x(i,1);
y0_Numerator =  (1/2)*(norm_pdf(xx,1,stan_dev));
y0_Denom = (1/2)*( norm_pdf(xx,1,stan_dev) ) + (1/2)*( norm_pdf(xx,2,stan_dev) );
y0_distrib(i,1) = y0_Numerator/y0_Denom;

end
for i = 1:100 
xx = x(i,1);
y1_Numerator =  (1/2)*(norm_pdf(xx,2,stan_dev)); 
y1_Denom = (1/2)*( norm_pdf(xx,1,stan_dev) ) + (1/2)*( norm_pdf(xx,2,stan_dev) );
y1_distrib(i,1) = y1_Numerator/y1_Denom;

end
% plot of special probability distribution for y = 0 
plot(x,y0_distrib)
hold on
% plot of special probability distribution for y = 1 
plot(x,y1_distrib)
title('Q4 part 3, Plot of x vs y')
xlabel('x') % x-axis label
ylabel('y') % y-axis label
legend('y0 distribution','y1 distribution')
```
