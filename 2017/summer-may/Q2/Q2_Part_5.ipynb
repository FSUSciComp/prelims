```matlab
N = 4;
m = 10;

trap_error = 0;

for j = 1 : m
    
   k = m*N;
   
  
   fun        = @(x)  exp(cos(x)).*cos(k*x);
   
   integral_a = integral(fun, 0, 2*pi);
   a_k        = (1/pi) * integral_a;
   trap_error = trap_error + 2 * pi * (a_k);
   
end

trap_error

m

N


scatter(trap_error,m)
```
