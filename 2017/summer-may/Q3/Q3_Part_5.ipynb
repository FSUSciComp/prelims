```matlab
%Student ID:5894371007989034

%Part 1
%Generate a Synthetic Data Set

N = 20;
t = zeros(N,1);
f = zeros(N,1);
a1 = 1.0;
a2 = 1.0;
b1 = -2.0;
b2 = -0.2;
t(1,1) = 0;
f(1,1) = a1*exp(b1*t(1,1)) + a2*exp(b2*t(1,1));
dt = 1;

for i = 2 : N
    
    t(i,1) = (i-1)*dt;
    f(i,1) = a1.*exp(b1*t(i,1)) + a2.*exp(b2*t(i,1));
  
end

'Part 1 results'
t
f

%Part 2

A = zeros(20,2);
y = f;
Y = zeros(2,1);

b1 = -2.0122;
b2 = -0.2019;

for i = 1 : 20
   
    A(i,1) = exp(b1*t(i,1));
    A(i,2) = exp(b2*t(i,1));
    
end

% AY = y
% Least Squares (AT) * (A) * (Y) = (AT) * (y)
AT = transpose(A);

Y = ((AT)*(A))\((AT)*(y));

a1 = Y(1,1);
a2 = Y(2,1);

'Part 5 results'

a1
a2



```
