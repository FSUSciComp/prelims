##### a)

The implementation of Newton's method is as following:


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2.0-4.0

def df(x):
    return 2.0*x

# newtons method:

x=2.5
err=np.zeros([7])
iteration=np.linspace(1, 7, num=7)

for i in range(7):
    x=x-f(x)/df(x)
    err[i]=f(x)
plt.semilogy(iteration, err) 
plt.show()
```

Newton's method has second order convergence rate. But from the figure above, we can see that the convergent rate is huge(about $log_210^{12}$). This is because the given function itself is second order, so the second order method can reach the true solution(which means no truncation error) in limited steps. To better estimate the convergence rate, higher order cost function is needed.

##### b)

Implementation of secant method is as following:


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2.0-4.0

# secant method:

x0=1.0
x1=2.5

err=np.zeros([7])
iteration=np.linspace(1, 7, num=7)

for i in range(7):
    temp=x1
    x1=x1-f(x1)*(x1-x0)/(f(x1)-f(x0))
    err[i]=np.absolute(f(x1))
    x0=temp
plt.semilogy(iteration, err) 
plt.show()
```

From the figure above, we can see that the convergent rate is about 10, which is much larger compared with the theoritical convergence rate of secant method(1.618), higher order cost function is needed in order to find out the true convergent rate via neumerical experiment.
