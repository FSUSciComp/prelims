The datastructure for the tree and how to construct tree using linked list is shown in the code as following:


```python
import sys

class Tree:
    
    #node for tree 
    class Node:
        def __init__(self, name= None, ancestor = None, left=None, right=None, data=None, index=None):
            
            self.name = name             #spicies name 
            self.ancestor = ancestor     # the node represent ancestor
            self.left = left             # left child 
            self.right = right           #right child
            self.data=data               #store the data
            self.index=index
            
        #getters for memeber datas
        def getName(self):         #get name of the species
            return self.name
            
        def getAncestor(self):     #get to the node of ancestor
            return self.ancestor         
  
        def getLeft(self):         #get the left child
            return self.left
            
        def getRight(self):        #get the right child
            return self.right  
            
        def getData(self):          #get the data
            return self.data      

        def getIndex(self):       
            return self.index 
        
        #setters for member datas
        def setName(self, name):           #set the species name
            self.name = name
            
        def setAncestor(self, ancestor):   #set the ancestor
            self.ancestor =ancestor

        def setLeft(self,left):            #set the left child
            self.left =left
            
        def setRight(self, right):         #set the right child
            self.right = right
            
        def setData(self,data):        #set the offspring number
            self.data = data
            
        def setIndex(self, index):       
            self.index= index
          
    #initialize tree as an empty node             
    def __init__(self):             
        self.root = Tree.Node()
        
def leaf(name, data): #return leaf node
    p=Tree()
    p.setName(name)
    p.setData(data)
    return p

def nonleaf(left, right): #return nonleaf node
    p=Tree()
    p.setLeft(left)
    p.setRight(right)
    return p

        
def constructTree(): #this could be determined by the input format, here is just a hard coded example of the given tree structure and data
    tree=Tree()
    tree.setLeft(leaf("Fish", "110110"))
    tree.setRight(nonleaf(leaf("Frog","001001"), nonleaf(leaf("Newt","100110"),nonleaf(leaf("snake","110000"),leaf("Toad","001110")))))
    return tree 
```
